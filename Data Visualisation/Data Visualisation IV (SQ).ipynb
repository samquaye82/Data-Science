```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
car_data = pd.read_csv('Cars93.csv')

group_by_Manufacturer = car_data.groupby(by=['Manufacturer'])

car_data_avg = round(group_by_Manufacturer.mean(),0)


data_of_interest = {'Audi': 2408,
                    'Huyundai': 2582,
                    'Suzuki': 3360,
                    'Toyota': 2691,
                    }

key_manufacturers = data_of_interest.keys()
Revs_values = data_of_interest.values()

df = pd.DataFrame({"Manufacturer": key_manufacturers, "Revs per mile": Revs_values})
ax = df.plot.box(column="Revs per mile", by="Manufacturer")
```

```python
# The manufacturer with the highest Revs per mile is Suzuki.
```

```python
# creating a dataframe for city and highway comparison
features_of_interest = pd.DataFrame({'City': car_data_avg['MPG.city'],
                                     'Highway': car_data_avg['MPG.highway'],
                                   }
                                   )

# plotting the histogram
features_of_interest.plot.hist()
```

```python
# From the data it is more fuel efficient to drive on the highway than it is to drive in the city.
```

```python
# creating a dataframe for wheelbase and turning circle comparison.
wheels_df = pd.DataFrame({'Wheelbase': car_data_avg['Wheelbase'],
                                     'Turning Circle': car_data_avg['Turn.circle'],
                                   }
                                   )

# plotting the line graph
wheels_df.plot.line(x='Turning Circle', y='Wheelbase')
```

```python
# The relationship between wheelbase and turning circle is they are postively correlated with each other. 
# As the wheelbase gets bigger, the turning circle gets bigger as well.
```

```python
# group by car type
group_by_type = car_data.groupby(by=['Type'])

car_data_avgs = round(group_by_type.mean(),0)
car_data_avgs
```

```python
# creating a dataframe for horsepower and length comparison.
type_df = pd.DataFrame({'Horsepower': car_data_avgs['Horsepower'],
                                     'Length': car_data_avgs['Length'],
                                   }
                                   )

# sorting the dataframe by length in ascending order
type_df.sort_values(by=['Length'], inplace=True)

# plotting the bar chart
type_df.plot.bar()
```

```python
# generally speaking a larger vehicle means a higher horsepower, 
# however there is a noticeable exception for sporty cars which have a significantly higher than expected horsepower.
```
